## ðŸŽ¯ What are **dispatch methods** in pandas?

**Dispatch methods** are **normal DataFrame or Series methods** that **pandas automatically sends ("dispatches") to each group** when you use `groupby`.

âœ… In other words:  
When you do something like `.sum()`, `.mean()`, `.min()`, `.max()`, `.count()`, etc. **after `groupby`**, pandas will automatically apply those methods **to each group separately**, **without you needing to write a custom function**.

---

# ðŸ“¦ Examples of dispatch methods:
- `.sum()`
- `.mean()`
- `.min()`
- `.max()`
- `.count()`
- `.first()`
- `.last()`
- `.std()`
- `.var()`
- `.median()`
- `.nunique()`
- etc.

**They are called "dispatch methods" because pandas "dispatches" the work to each group automatically.**

---

### ðŸ›  How dispatch methods behave on `groupby`?

When you use a dispatch method after `groupby`:

1. **Pandas splits the DataFrame** into groups.
2. **It runs the method** separately on **each group**.
3. **It combines** the results back into a single new DataFrame or Series.

---

#### ðŸ“š Let's see an example:

```python
import pandas as pd

data = pd.DataFrame({
    'Category': ['Furniture', 'Furniture', 'Technology', 'Office Supplies', 'Technology', 'Office Supplies'],
    'Sales': [200, 300, 400, 150, 600, 250]
})

# Group by Category and find the sum of Sales
grouped_sales = data.groupby('Category')['Sales'].sum()
print(grouped_sales)
```

ðŸ”µ Output:

```
Category
Furniture           500
Office Supplies     400
Technology         1000
Name: Sales, dtype: int64
```

âœ… `.sum()` was dispatched automatically to each group!

- **Furniture** â†’ 200 + 300 = 500
- **Office Supplies** â†’ 150 + 250 = 400
- **Technology** â†’ 400 + 600 = 1000

---

### ðŸ“¢ Summary:  
| Feature | What it means |
|:---|:---|
| **Dispatch Methods** | Normal DataFrame/Series methods that pandas automatically runs on each group |
| **Behavior** | Splits â†’ Applies the method on each group â†’ Combines the results |
| **Examples** | `.sum()`, `.mean()`, `.count()`, `.min()`, `.max()`, `.std()`, etc. |

---

### ðŸš€ Quick difference:

| Concept | Behavior |
|:---|:---|
| `.apply()` | You must manually tell pandas what custom logic to apply on each group. |
| **Dispatch method** (e.g., `.sum()`) | Pandas **already knows** what to do â€” no custom function needed. |

---

### âš¡ One-liner to remember:

> **If pandas already knows what you want (sum, mean, count, etc.), use dispatch methods. Otherwise, use `apply()`.**

## ðŸŽ¯ What `.agg()` can take?

When you use `.agg()` after a `groupby` (or even directly on a DataFrame), you can pass in:

| What you pass | Example | What it means |
|:---|:---|:---|
| **Single function name as string** | `'mean'`, `'sum'`, `'min'`, `'max'`, `'count'` | Apply one function |
| **List of function names** | `['mean', 'sum']` | Apply multiple built-in functions |
| **Custom function** | `lambda x: x.max() - x.min()` | Your own function |
| **List of custom + built-in functions** | `['mean', lambda x: x.max() - x.min()]` | Mix and match |
| **Dictionary** | `{'Marks': ['mean', 'max'], 'Age': 'min'}` | Apply specific functions to specific columns |

---

## ðŸ› ï¸ Examples

### 1. **Single built-in function**
```python
df.groupby('Class')['Marks'].agg('mean')
```
ðŸ‘‰ Find average marks per class.

---

### 2. **List of multiple built-in functions**
```python
df.groupby('Class')['Marks'].agg(['mean', 'sum', 'max'])
```
ðŸ‘‰ Find mean, sum, and max marks per class.

---

### 3. **Custom function**
```python
df.groupby('Class')['Marks'].agg(lambda x: x.max() - x.min())
```
ðŸ‘‰ Find the range (max - min) marks per class.

---

### 4. **Mixing built-in and custom functions**
```python
df.groupby('Class')['Marks'].agg(['mean', lambda x: x.max() - x.min()])
```
ðŸ‘‰ Get mean and range together.

---

### 5. **Using dictionary (different columns different functions)**

Suppose you have more columns:
```python
data = {
    'Class': ['A', 'A', 'B', 'B', 'B', 'C'],
    'Marks': [85, 90, 78, 82, 89, 95],
    'Age': [14, 15, 14, 15, 14, 16]
}

df = pd.DataFrame(data)
```

You can do:

```python
df.groupby('Class').agg({
    'Marks': ['mean', 'max'],
    'Age': 'min'
})
```

ðŸ‘‰ Here:
- For `Marks`, it finds mean and max.
- For `Age`, it finds minimum.

---

## ðŸ”¥ In short:
- `.agg()` is like a **Swiss army knife** ðŸ› ï¸ â€” you can customize **what** and **how many things** you want to calculate.
- You can pass strings, lists, functions, and even dictionaries!

---

## ðŸ”¥ Quick thumb rule:

> **"If it's simple math/statistics âž” use `agg()`."**  
> **"If it's complex decisions/transformations âž” use `apply()`."**

````python
# Say, I want to categorize sales for each category as Above Avergae abd Below Average based on whether they are greater than average sales for that category

# Define a custom function to categorize sales within each group
def categorize_sales(group):
    # Calculate the average sale for the current group (each 'Category')
    avg_sale = group['Sales'].mean()
    
    # For each Sale in the group, label it as 'Above Average' or 'Below Average'
    # based on whether it is greater than the group's average
    return group['Sales'].apply(
        lambda x: 'Above Average' if x > avg_sale else 'Below Average'
    )

# Group the data by 'Category'
# Then apply the 'categorize_sales' function to each group
# 'include_groups=False' ensures that the 'Category' column is NOT included inside each group (future-proof)
sales_category = data.groupby('Category').apply(categorize_sales, include_groups=False)

# View the resulting Series
sales_category
````

````python
# Combine with original DataFrame
df_clean = data.copy(deep=True)
df_clean['Sales_Category'] = sales_category.values  # .values to align the index
df_clean.head()
````
